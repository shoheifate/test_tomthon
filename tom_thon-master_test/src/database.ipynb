```python
#https://qiita.com/saira/items/e08c8849cea6c3b5eb0c
import sqlite3
import pandas as pd
import os

dest = 'databases'
if not os.path.exists(dest):
    os.makedirs(dest)

# pandascsv
# csv1label, 2wind_speed, 3temperature, 4precipitation_amount, 
#5wave_height
df = pd.read_csv("isigaki_miyako_iriomote_wavedata/isigaki_test.csv", header=None)

# csv filecolumn
# pandas1

df.columns = ["label","wind_speed","temperature","precipitation_amount","wave_height"]

dbname = os.path.join(dest,'TEST.db')

conn = sqlite3.connect(dbname)
cur = conn.cursor()

# dbnamesamplecsvsql
# if_existsexpense
df.to_sql('sample', conn, if_exists='replace')

# 1
select_sql = 'SELECT * FROM sample'
for row in cur.execute(select_sql):
    print(row)

cur.close()
conn.close()
```
